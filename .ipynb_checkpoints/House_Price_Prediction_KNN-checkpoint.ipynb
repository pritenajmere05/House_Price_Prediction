```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error

import seaborn as sns
import matplotlib.pyplot as plt
```

### Data Gathering

```python
df = pd.read_csv('Housing_V1.csv')
df
```

### Model Training

```python
x = df.drop('price' , axis = 1)
y = df['price']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 32)
```

```python
knn_reg = KNeighborsRegressor()
```

```python
knn_reg.fit(x_train , y_train)
```

```python
knn_reg.score(x_train , y_train)
```

```python
knn_reg.score(x_test , y_test)
```

```python
knn_reg = KNeighborsRegressor(p=1)
```

```python
knn_reg.fit(x_train , y_train)
```

```python
knn_reg.score(x_train , y_train)
```

```python
knn_reg.score(x_test , y_test)
```

### Model Evaluation

```python
k_values = np.arange(2,16)
train_acc_list = []
test_acc_list = []

for k in k_values:
    knn_reg = KNeighborsRegressor(n_neighbors=k , p=1)
    knn_reg.fit(x_train,y_train)
    train_acc = knn_reg.score(x_train,y_train)
    train_acc_list.append(train_acc)
    
    test_acc = knn_reg.score(x_test,y_test)
    test_acc_list.append(test_acc)
    
plt.plot(k_values,train_acc_list)
plt.plot(k_values,test_acc_list)
plt.title("Comparision between train and Test Accuracies")
plt.xlabel("K Values")
plt.ylabel("Accuracies")
```

```python
k_values = np.arange(2,16)
train_acc_list = []
test_acc_list = []

for k in k_values:
    knn_reg = KNeighborsRegressor(n_neighbors=k , p=2)
    knn_reg.fit(x_train,y_train)
    train_acc = knn_reg.score(x_train,y_train)
    train_acc_list.append(train_acc)
    
    test_acc = knn_reg.score(x_test,y_test)
    test_acc_list.append(test_acc)
    
plt.plot(k_values,train_acc_list)
plt.plot(k_values,test_acc_list)
plt.title("Comparision between train and Test Accuracies")
plt.xlabel("K Values")
plt.ylabel("Accuracies")
```

### Hyperparameter Tunning

```python
knn_reg = KNeighborsRegressor()
parameters = {'n_neighbors' : np.arange(2,30) , 'p' : [1,2] }

gscv_knn_reg = GridSearchCV(knn_reg , parameters)
gscv_knn_reg.fit(x_train,y_train)
```

```python
gscv_knn_reg.best_estimator_
```

```python
knn_reg = gscv_knn_reg.best_estimator_
knn_reg.fit(x_train,y_train)
```

```python
knn_reg.score(x_train,y_train)
```

```python
knn_reg.score(x_test,y_test)
```

### Standardizaion

```python
std_scalar = StandardScaler()
```

```python
std_scalar.fit(x)
array = std_scalar.transform(x)
x_df = pd.DataFrame(array , columns = x.columns)
x_df
```

```python
y = df['price']

x_train , x_test , y_train , y_test = train_test_split(x_df , y , test_size = 0.2 , random_state = 32)
```

```python
k_values = np.arange(2,16)
train_acc_list = []
test_acc_list = []

for k in k_values:
    knn_reg = KNeighborsRegressor(n_neighbors=k , p=1)
    knn_reg.fit(x_train,y_train)
    train_acc = knn_reg.score(x_train,y_train)
    train_acc_list.append(train_acc)
    
    test_acc = knn_reg.score(x_test,y_test)
    test_acc_list.append(test_acc)
    
plt.plot(k_values,train_acc_list)
plt.plot(k_values,test_acc_list)
plt.title("Comparision between train and Test Accuracies")
plt.xlabel("K Values")
plt.ylabel("Accuracies")
```

```python
k_values = np.arange(2,16)
test_acc_list = []
train_acc_list = []

for k in k_values:
    knn_reg = KNeighborsRegressor(n_neighbors=k, p = 2)
    knn_reg.fit(x_train, y_train)
    train_acc = knn_reg.score(x_train, y_train)
    train_acc_list.append(train_acc)
    
    test_acc = knn_reg.score(x_test, y_test)
    test_acc_list.append(test_acc)
    # print(k, train_acc, test_acc)
    
plt.plot(k_values, test_acc_list)
plt.plot(k_values, train_acc_list)
plt.title("Comparision between train and Test Accuracies")
plt.xlabel("K Values")
plt.ylabel("Accuracies")
```

```python
knn_reg = KNeighborsRegressor()
```

```python
knn_reg.fit(x_train,y_train)
```

```python
#Training data evaluation

y_train_pred = knn_reg.predict(x_train)

mae = mean_absolute_error(y_train , y_train_pred)
print("Mean absolute error : ",mae)

mse = mean_squared_error(y_train , y_train_pred)
print("Mean squared error : ",mse)

r2 = r2_score(y_train , y_train_pred)
print("R2 Score : ",r2)
```

```python
#Testing data evaluation

y_test_pred = knn_reg.predict(x_test)

mae = mean_absolute_error(y_test , y_test_pred)
print("Mean absolute error : ",mae)

mse = mean_squared_error(y_test , y_test_pred)
print("Mean squared error : ",mse)

r2 = r2_score(y_test , y_test_pred)
print("R2 Score : ",r2)
```
